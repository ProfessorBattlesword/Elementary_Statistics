##### Objective 1: How to find the range

##### Range

* The **range** of a data set is the difference between the largest and smallest data entries (quantitative data only)
    * Advantage: easy to compute
    * Disadvantage: only uses two data entries

##### Objective 2: How to find the variance and standard deviation of a population and of a sample

##### Deviation and Population Variance

* The **deviation** of an entry $x$ in a data set is the difference between itself and the mean
    * $\text{Deviation of } x = x - \mu$
    * The sum of all deviations for a data set is *always* zero
    * Therefore, the average of the deviations for a data set is *always* zero
    * Therefore, squaring the deviations and then averaging is more meaningful
* The **population variance** of a population data set is given by $$\sigma^2 = \cfrac{\sum{(x - \mu)^2}}{N},$$ where $N$ is the population size
    * Disadvantage: units of variance are squared

##### Population Standard Deviation

* The **population standard deviation** of a population data set is given by $$\sigma = \sqrt{\cfrac{\sum{(x - \mu)^2}}{N}},$$ where $N$ is the population size
    * Measures the variation of the data about the mean and has the same units as the original data set
    * $\sigma \geq 0$ and when $\sigma = 0$ all data entries are the same (the data does not vary whatsoever)
    * As data entries diverge from the mean the standard deviation increases (data entries are more spread out)

**Note:** The same rounding rule as used for calculating the mean is used for standard deviation and variance (i.e. round to one more decimal place than the original data provided/collected).

##### Sample Variance and Sample Standard Deviation

* The **(unbiased) sample variance** is given by $$s^2 = \cfrac{\sum(x - \bar{x})^2}{n - 1}$$

* The **sample standard deviation** is given by $$s = \sqrt{\cfrac{\sum{(x - \bar{x})^2}}{n - 1}}$$

**Note:** The $n - 1$ in place of $N$ for sample standard deviation/variance is due to theoretical reasons regarding random samples that we will not get into. Also of note, there are two types of variance/standard deviations: biased and unbiased. We will consider only the unbiased sample standard deviation/variance in this course.

##### Objective 3: How to use the Empirical Rule and Chebychev's Theorem to interpret standard deviation

##### Interpreting Standard Deviation

* In general, standard deviation tells us how much our data is spread out *relative to the mean.*
* The larger the standard deviation the more our data is spread away from the mean.
* If we have a small standard deviation then our data is "bunched" up toward the mean.
* Data entries that lie *more than two* standard deviations away from the mean are considered **unusual.**
* Data entries that lie *more than three* standard deviations away from the mean are considered **very unusual.**

```python
import warnings

warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

```python
boring_data = [5,5,5,5,5,5,5,5]
symmetric_data = [3,4,4,5,5,5,5,6,6,7]
bimodal_data = [2,2,2,3,7,8,8,8]

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(25,10))

##### No variance
sns.histplot(boring_data, kde=False, discrete=True, ax=ax1, stat='frequency')
ax1.set_xticks([1,2,3,4,5,6,7,8,9])
ax1.set_title(r'Distribution with $\bar{x} = 5$ and $s = 0$', fontsize=30)

##### low spread
sns.histplot(symmetric_data, kde=False, discrete=True, ax=ax2, stat='frequency')
ax2.set_xticks([1,2,3,4,5,6,7,8,9])
ax2.set_yticks([1,2,3,4,5,6,7,8])
ax2.set_title(r'Distribution with $\bar{x} = 5$ and $s = 1.2$', fontsize=30)

##### large spread
sns.histplot(bimodal_data, kde=False, discrete=True, ax=ax3, stat='frequency')
ax3.set_xticks([1,2,3,4,5,6,7,8,9])
ax3.set_yticks([1,2,3,4,5,6,7,8])
ax3.set_title(r'Distribution with $\bar{x} = 5$ and $s = 3.0$', fontsize=30)

fig.tight_layout()
```

##### Empirical Rule

The **empirical rule** is a general rule for *symmetric* or *approximately symmetric* data that aids in estimating where most of the data lies based on standard deviation.

* For data sets (population or sample) that are approximately *symmetric* and *bell-shaped:*
    * About $68\%$ of the data lies within one standard deviation of the mean
    * About $95\%$ of the data lies within two standard deviations of the mean
    * About $99.7\%$ of the data lies within two standard deviations of the mean
    * See the graph below for a visual depiction of the empirical rule

```python
import scipy.stats as stats

sigma = 50
mu = 350

x = np.arange(mu-4*sigma,mu+4*sigma+.1,.1)
y = stats.norm.pdf(x,mu,sigma)

fig, ax = plt.subplots(figsize=(25,10))

ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=mu, ymax=.95, color='k')
ax.axvline(x=mu-sigma, ymax=.6, color='k')
ax.axvline(x=mu+sigma, ymax=.6, color='k')
ax.axvline(x=mu-2*sigma, ymax=.17, color='k')
ax.axvline(x=mu+2*sigma, ymax=.17, color='k')
ax.axvline(x=mu-3*sigma, ymax=.07, color='k')
ax.axvline(x=mu+3*sigma, ymax=.07, color='k')
ax.annotate('$34\%$',(mu+15,0.002), fontsize=35)
ax.annotate('$34\%$',(mu-35,0.002), fontsize=35)
# ax.annotate('$13.5\%$',(mu+sigma+15,0.001), fontsize=35)
# ax.annotate('$13.5\%$',(mu-sigma-35,0.001), fontsize=35)
ax.annotate('$13.5\%$', xy=(mu+sigma+15, 0.001), xytext=(mu+sigma+20, 0.004),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=35)
ax.annotate('$13.5\%$', xy=(mu-sigma-25, 0.001), xytext=(mu-sigma-45, 0.004),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=35)
ax.annotate('$2.35\%$', xy=(mu+2*sigma+15, 0.00005), xytext=(mu+2*sigma+20, 0.002),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=35)
ax.annotate('$2.35\%$', xy=(mu-2*sigma-15, 0.00005), xytext=(mu-2*sigma-35, 0.002),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=35)
ax.annotate('$0.15\%$', xy=(mu+3*sigma+15, 0.00005), xytext=(mu+3*sigma+25, 0.002),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=35)
ax.annotate('$0.15\%$', xy=(mu-3*sigma-15, 0.00005), xytext=(mu-3*sigma-35, 0.002),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=35)
ax.fill_between(x,0,y,color='green')
ax.set_title('Empirical Rule', fontsize=35)
ax.set_xticks([mu-3*sigma,mu-2*sigma,mu-sigma,
               mu,mu+sigma,mu+2*sigma,mu+3*sigma])
ax.set_xticklabels(['$\mu - 3 \cdot \sigma$','$\mu - 2 \cdot \sigma$',
                    '$\mu - \sigma$', '$\mu$', '$\mu + \sigma$',
                    '$\mu + 2 \cdot \sigma$', '$\mu + 3 \cdot \sigma$'],
                   fontsize=25, rotation=45)
ax.set_yticks([])
ax.set_xlim((mu-4*sigma,mu+4*sigma))
ax.set_ylim=(0,stats.norm.pdf(mu,mu,sigma)+.01)
```

##### Chebychev's Theorem

* The portion of *any* data set lying within $k$ standard deviations ($k>1$) of the mean is at least $1 - \cfrac{1}{k^2}$.
    * If $k = 2$, then in *any* data set at least $75\%$ of the data lie within 2 standard deviations of the mean
    * If $k = 3$, then in *any* data set at least $88.9\%$ of the data lie within 3 standard deviations of the mean

##### Objective 4: How to use the coefficient of variation to compare variation in different datasets

##### Coefficient of Variation

* The **coefficient of variation (CV)** of a data set describes the standard deviation as a percent of the mean.

* Population: $$CV = \cfrac{\sigma}{\mu} \cdot 100\%$$

* Sample: $$CV = \cfrac{s}{\bar{x}} \cdot 100\%$$

##### Example 7

The table below shows the population heights (in inches) and weights (in pounds) of the members of a basketball team. Find the coefficient of variation for the heights and weights. Then compare the results.

|Heights|Weights|
|:-:|:-:|
|72|180|
|74|168|
|68|225|
|76|201|
|74|189|
|69|192|
|72|197|
|79|162|
|70|174|
|69|171|
|77|185|
|73|210|

##### Solution:

* We are dealing with all members of a basketball team and therefore we are working with a population

* For the heights:
    * Pop st dev: $\sigma \approx 3.3$
    * Pop mean: $\mu \approx 72.8$
    * Pop CV: $CV = \cfrac{\sigma}{\mu} \cdot 100\% = \cfrac{3.3}{72.8} \cdot 100\% \approx 4.53\%$
* For the weights:
    * Pop st dev: $\sigma \approx 17.7$
    * Pop mean: $\mu \approx 187.8$
    * Pop CV: $CV = \cfrac{\sigma}{\mu} \cdot 100\% = \cfrac{17.7}{187.8} \cdot 100\% \approx 9.42\%$

Comparing results there is a higher percentage of variation in weight meaning there is more variance in how much players weigh than there is in how much players vary in height.
