##### Stem-and-Leaf Plot

* A **stem-and-leaf plot** splits each number in a data set into a **stem** (usually the leftmost digit(s) of a number) and a **leaf** (usually the rightmost digit).

* Stem-and-leaf plots show a similar pattern as a histogram.

* Advantage: shows the *original data.*

##### Dot Plot

* A **dot plot** consists of each data entry plotted using a point above a horizontal axis

* Dot plots are useful in seeing how data are distributed, determining specific data entries, and identifying unusual data entries

##### Example 4

Construct a dot plot for the following data.

* Number of text messages sent last week by the cell phone users on one floor of a college dormatory:

|||||||||||
|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
|155|159|144|129|105|145|126|116|130|114|
|122|112|112|142|126|118|118|108|122|121|
|109|140|126|119|113|117|118|108|122|121|
|139|139|122|78|133|126|123|145|121|134|
|124|119|132|133|124|129|112|126|148|147|

##### Solution:

```python
data = [155,159,144,129,105,
     145,126,116,130,114,
     122,112,112,142,126,
     118,118,108,122,121,
     109,140,126,119,113,
     117,118,108,122,121,
     139,139,122,78,133,
     126,123,145,121,134,
     124,119,132,133,124,
     129,112,126,148,147]

bins = np.arange(min(data),max(data)+1)-0.5

hist, edges = np.histogram(data, bins=bins)

y = np.arange(1,hist.max()+1)
x = np.arange(min(data),max(data))
X,Y = np.meshgrid(x,y)

fig, ax = plt.subplots(figsize=(25,10))

plt.style.use('fivethirtyeight')

ax.scatter(X,Y, c=Y<=hist, cmap="Greys")
ax.set_xticks(x)
ax.set_xticklabels(x,rotation=90,fontsize=15)
ax.set_yticks(y)
ax.set_yticklabels(y,fontsize=15)

ax.set_title('Number of Text Messages Sent', fontsize = 25)
ax.set_xlabel('Number of Text Messages', fontsize = 25)
ax.set_ylabel('Frequency', fontsize = 25)
```

##### Objective 2: How to graph and interpret qualitative datasets using pie charts and Pareto charts

##### Graphing Qualitative Data Sets

* Pie Chart

##### Pie Chart

* A **pie chart** is a circle that is divided into sections that represent categories.

##### Example 6

Represent the following data using a pie chart.

|Type of Degree|Number (in thousands)|
|:-:|:-:|
|Associate's|455|
|Bachelor's|1051|
|Master's|330|
|Doctoral|104|

##### Solution:

```python
import pandas as pd

df = pd.DataFrame({'Type of Degree': ['Associate\'s', 'Bachelor\'s', 'Master\'s', 'Doctoral'],'Number (in thousands)': [455, 1051, 330, 104]})
df.set_index('Type of Degree', inplace=True)

ax = df.plot(kind='pie',y='Number (in thousands)',autopct='%1.1f%%',shadow=True,startangle=90,fontsize=25, figsize=(25,10))

ax.legend(loc=(1,.85))
ax.set_ylabel('')
ax.set_title('Types of Degrees')
```

##### Pareto Chart

* A **Pareto Chart** is a vertical bar graph in which the height of each bar represents frequency or relative frequency in decreasing order.

##### Example 7

A survey was conducted for data scientists to determine subjects of greatest interest. Represent the "Very interested" column using a Pareto chart.

||Very interested|Somewhat interested|Not interested|
|:-:|:-:|:-:|:-:|
|Big Data (Spark/Hadoop)|1332|729|127|
|Data Analysis/Statistics|1688|444|60|
|Data Journalism|429|1081|610|
|Data Visualization|1340|734|102|
|Deep Learning|1263|770|136|
|Machine Learning|1629|477|74|

##### Solution:

```python
df = pd.read_csv('Topic_Survey_Assignment.csv')

df.rename(columns={'Unnamed: 0': ''}, inplace=True)
df.set_index('', inplace=True)
df.sort_values(['Very interested'], ascending=False, inplace=True)

ax = df['Very interested'].plot(kind='bar',figsize=(25,10),fontsize=25)
ax.set_title('Data Science Topics (Very Interested)')
```

##### Scatter Plot

* A **scatter plot** is the plotting of ordered pairs in the coordinate plane (plotting paired ata in $xy$-plane)

##### Example 8

Use a scatter plot to describe the following data.

* Hourly Wages:

|Hours|Hourly Wage|
|:-:|:-:|
|33|12.16|
|37|9.98|
|34|10.79|
|40|11.71|
|35|11.80|
|33|11.51|
|40|13.71|
|33|12.05|
|28|10.54|
|45|10.33|
|37|11.57|
|28|10.17|

##### Solution:

```python
hours = [33,37,34,40,35,33,40,33,28,45,37,28]
hours_worked = [12.16,9.98,10.79,11.71,11.80,11.51,13.71,12.05,10.54,10.33,11.57,10.17]

fig, ax = plt.subplots(figsize=(25,10))

ax.scatter(hours, hours_worked)

ax.set_xticks(np.arange(min(hours),max(hours)+1))

ax.set_title('Hourly Wages', fontsize = 25)
ax.set_xlabel('Hours Worked', fontsize = 25)
ax.set_ylabel('Wage Earned', fontsize = 25)
```

##### Time Series Chart

* A **time series chart** is a line graph where the horizontal axis represents fixed intervals of time and the vertical axis represents some quantitative measure over the time period.

##### Example 9

Plot the number of subscribers through 2002 to 2012 using a time series chart for the following data. Also plot the average bill from year 2002 to 2012.

* List of cell phone subscribers (in millions) and subscriber's average local monthly bills for service (in dollars) for the years 2002 through 2012:

|Year|Subscribers (in millions)|Average bill (in dollars)|
|:-:|:-:|:-:|
|2002|134.6|47.42|
|2003|148.1|49.46|
|2004|169.5|49.49|
|2005|194.5|49.52|
|2006|219.7|49.20|
|2007|243.4|49.94|
|2008|262.7|48.54|
|2009|276.6|49.57|
|2010|292.8|47.47|
|2011|306.3|47.23|
|2012|321.7|47.16|

##### Solution:

```python
years = np.arange(2002,2013)
subs = [134.6,148.1,169.5,194.5,219.7,243.4,262.7,276.6,292.8,306.3,321.7]
avg_bill = [47.42,49.46,49.49,49.52,49.20,49.94,48.54,49.57,47.47,47.23,47.16]

fig, ax = plt.subplots(figsize=(25,10))

ax.plot(years, subs, color='blue', marker='o', markersize=10)

ax.set_xticks(np.arange(min(years),max(years)+1))

ax.set_title('Cell Phone Subscribers', fontsize = 25)
ax.set_xlabel('Year', fontsize = 25)
ax.set_ylabel('Subscribers (in millions)', fontsize = 25)
```
