##### Objective 1: How to construct a frequency distribution including limits, midpoints, relative frequencies, cumulative frequencies, and boundaries

##### Frequency Distribution

* A **frequency distribution** is a table that shows **classes** or **intervals** of data entries with a count of the number of entries in each class.
* The **frequency**, $f$, of class is the number of data entries in the class.
* Frequency distributions are useful when needing to organize large data sets
* ##### Example:

| Class | Frequency, $f$ |
|:-:|:-:|
| 1-5 | 5 |
| 6-10 | 8 |
| 11-15 | 6 |
| 16-20 | 8 |
| 21-25 | 5 |
| 26-30 | 4 |

##### Steps for Constructing a Frequency Distribution from Data

1. Decide the number of classes to include (usually between 5 and 20 classes)
2. Find the class width (*always* round up): $\textbf{class width} = \cfrac{\textbf{range}}{\textbf{# of classes}}$.
3. Find class limits:
    * Use the *lowest* data entry as the lower class limit of the first class
    * To find the remaining lower class limits add the class width to the preceding lower class limit until you have the desired number of classes
    * To find the first upper class limit do the following (to avoid data overlapping into separate classes): $\textbf{first upper class limit} = \textbf{lowest class limit} + (\textbf{class width} - 1)$
    * To find the remaining upper class limits add the class width to the preceding upper class limit until you have an upper class limit for each class
4. Count the data entries that fall into each class

##### Characteristics of Frequency Distributions

* The **midpoint** of a class is given by the formula $$\textbf{midpoint} = \cfrac{\textbf{lower class limit} + \textbf{upper class limit}}{2}.$$
* The **relative frequency** of a class is the portion, or percentage, of the data that falls in that class. This can be found using the formula $$\textbf{relative frequency} = \cfrac{\textbf{class frequency}}{\textbf{sample size}} = \cfrac{f}{n}.$$
* The **cumulative frequency** of a class is the sum of the frequencies of that class and all previous classes. The last class is equal to the sample size $\left( \sum f \right) = n.$

##### Frequency Histogram

* A **frequency histogram** is a bar graph that represents the frequency distribution of a data set. A histogram has the following properties:
    * The horizontal scale is *quantitative* and measures the data entries.
    * The vertical scale measures the *frequencies* of the classes.
    * Consecutive bars must touch.
* **Class boundaries** are the numbers that separate classes *without* forming gaps between them (since consecutive bars must touch).
* To find class boundaries (of integer data): subtract 0.5 from each lower class limit (this gives the **lower class boundaries**) and add 0.5 to each upper class limit (this gives the **upper class boundaries**).

##### Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')

x = np.array([128,100,180,150,200,
              90,340,105,85,270,
              200,65,230,150,150,
              120,130,80,230,200,
              110,126,170,132,140,
              112,90,340,170,190])
fig, ax = plt.subplots(figsize=(25,10))

ax.hist(x,bins=7,edgecolor='k')
ax.set_xticks(np.arange(64.5,345,40))
ax.set_yticks(np.arange(0,10,1))

ax.set_title('Prices of Retail Gadgets',fontsize=35)
ax.set_xlabel('Price (in dollars)',fontsize=35)
ax.set_ylabel('Frequency',fontsize=35)
```

##### Relative Frequency Histogram

* A **relative frequency histogram** has the same shape and the same horizontal scale as the corresponding frequency histogram, but the *vertical scale* of the corresponding frequency histogram is replaced with the *relative frequency*.

```python
x = np.array([128,100,180,150,200,
              90,340,105,85,270,
              200,65,230,150,150,
              120,130,80,230,200,
              110,126,170,132,140,
              112,90,340,170,190])

fig, ax = plt.subplots(figsize=(25,10))

ax.hist(x,bins=7,weights=np.zeros_like(x) + 1. / x.size,edgecolor='k')
ax.set_xticks(np.arange(64.5,345,40))
# ax.set_yticks(np.arange(0,.2,.1))

ax.set_title('Prices of Retail Gadgets',fontsize=35)
ax.set_xlabel('Price (in dollars)',fontsize=35)
ax.set_ylabel('Relative Frequency',fontsize=35)
```

##### Frequency Polygon

* A **frequency polygon** is a line graph that emphasizes the *continuous* change in frequencies
    * Use the midpoint of each class along with the corresponding frequency to plot points
    * Subtract the class width from the lowest midpoint and plot a point with frequency *zero*.
    * Add the class width to the highest midpoint and plot a point with frequency *zero*.
    * Connect the points with line segments

```python
x = np.array([128,100,180,150,200,
              90,340,105,85,270,
              200,65,230,150,150,
              120,130,80,230,200,
              110,126,170,132,140,
              112,90,340,170,190])
mid = [44.5, 84.5, 124.5, 164.5, 204.5, 244.5, 284.5, 324.5, 364.5]
freq = [0,6,9,6,4,2,1,2,0]

fig, ax = plt.subplots(figsize=(25,10))

ax.plot(mid,freq,marker='o',ms=15,markerfacecolor='r',color='k')
ax.set_xticks(mid)
ax.set_yticks(np.arange(0,10,1))

ax.set_title('Prices of Retail Gadgets',fontsize=35)
ax.set_xlabel('Price (in dollars)',fontsize=35)
ax.set_ylabel('Relative Frequency',fontsize=35)
```

##### Ogive

* An **ogive** is a line graph that displays the *cumulative frequency* of each class at its *upper class boundary*. The upper class boundaries are marked on the horizontal axis, and the cumulative frequencies are marked on the vertical axis.

```python
mid = [64.5, 104.5, 144.5, 184.5, 224.5, 264.5, 304.5, 344.5, 384.5]
cumul_freq = [0,6,15,21,25,27,28,30,30]

fig, ax = plt.subplots(figsize=(25,10))

ax.plot(mid,cumul_freq,marker='o',ms=15,markerfacecolor='r',color='k')
ax.set_xticks(mid)
ax.set_yticks(np.arange(0,30,1))

ax.set_title('Prices of Retail Gadgets',fontsize=35)
ax.set_xlabel('Price (in dollars)',fontsize=35)
ax.set_ylabel('Relative Frequency',fontsize=35)
```
