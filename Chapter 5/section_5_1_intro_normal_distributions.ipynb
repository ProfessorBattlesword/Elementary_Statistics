##### Objective 1: Interpreting Graphs of Normal Probability Distributions

##### Normal Distribution

A **normal distribution** is a *continuous* probability distribution for a random variable $x$ which has the following properties:

* The mean, median, and mode are all equal.

* The normal curve is bell-shaped and is symmetric about the mean.

* The total area under the curve is equal to 1.

* The curve approaches, but never touches, the $x$-axis as it extends farther and farther away from the mean.

* Between $x=\mu - \sigma$ and $x=\mu + \sigma$ the graph curves downward (cupped downward).

* The graph curves upward to the left of $x = \mu - \sigma$ and to the right of $x = \mu + \sigma$

* The points where the curvature changes are called <em><b>inflection points</b></em>.

* The graph of the normal distribution is called a <em><b>normal curve.

##### Example 1

Normal curve with $\mu = 3.5$ and $\sigma = 0.7$:

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def P(x,sigma,mu):
    return 1/(sigma*np.sqrt(2*np.pi)*np.exp((x-mu)**2/(2*sigma**2)))

sigma = .7
mu = 3.5
x = np.arange(mu-4*sigma,mu+4*sigma+.1,.1)
y = P(x,sigma,mu)

fig, ax = plt.subplots(figsize=(25,10))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='green')
ax.annotate('Inflection Point',
            xy=(mu+sigma, P(mu+sigma,sigma,mu)),
            xytext=(0.65, 0.6),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            color='red',
           fontsize=25)
ax.annotate('Inflection Point',
            xy=(mu-sigma, P(mu-sigma,sigma,mu)),
            xytext=(.3, 0.6),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='right',
            verticalalignment='bottom',
            color='red',
           fontsize=25)
ax.annotate('Mean',
            xy=(mu, P(mu,sigma,mu)),
            xytext=(.5, 0.8),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='blue', shrink=0.05),
            horizontalalignment='right',
            verticalalignment='bottom',
            color='blue',
           fontsize=25)
ax.set_title('Normal Curve', fontsize=25)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_xticklabels(np.round(np.arange(mu-3*sigma,mu+3*sigma+1,sigma), 2), fontsize=25)
ax.set_yticks(np.arange(0,1.1,.1))
ax.set_yticklabels(np.round(np.arange(0,1.1,.1), 2), fontsize=25)
ax.set(xlim=(-1.5,9),ylim=(0,1))
```

##### Objective 2: Finding Areas Under the Standard Normal Curve

##### Standard Normal Distribution

The <em><b>Standard Normal Distribution</b></em> is a normal distribution such that the <em>mean</em> is zero ($\mu = 0$) and the <em>standard deviation</em> is one ($\sigma = 1$). Standard Normal Distribution uses the standard score or $z$-score as inputs instead of $x$ values.

An example of the Standard Normal Distribution is shown below (this is one <em>unique</em> distribution):

```python
sigma = 1
mu = 0
x = np.arange(mu-4*sigma,mu+4*sigma+.1,.1)
y = P(x,sigma,mu)

fig, ax = plt.subplots(figsize=(25,15))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='green')
ax.annotate('Inflection Point',
            xy=(mu+sigma, P(mu+sigma,sigma,mu)),
            xytext=(0.7, 0.6),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            color='red',
           fontsize=25)
ax.annotate('Inflection Point',
            xy=(mu-sigma, P(mu-sigma,sigma,mu)),
            xytext=(.3, 0.6),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='right',
            verticalalignment='bottom',
            color='red',
           fontsize=25)
ax.annotate('Mean',
            xy=(mu, P(mu,sigma,mu)),
            xytext=(.5, 0.7),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='blue', shrink=0.05),
            horizontalalignment='right',
            verticalalignment='bottom',
            color='blue',
           fontsize=25)
ax.annotate('Total (shaded) area = 1',
            xy=(mu-.75*mu, P(mu,sigma,mu)-.75*P(mu,sigma,mu)),
            xytext=(.95, 0.25),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='green', shrink=0.05),
            horizontalalignment='right',
            verticalalignment='bottom',
            color='green',
           fontsize=25)
ax.set_title('Standard Normal Curve', fontsize = 25)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_xticklabels(np.arange(mu-3*sigma,mu+3*sigma+1,sigma), fontsize=25)
ax.set_yticks(np.arange(0,1.1,.1))
ax.set_yticklabels(np.round(np.arange(0, 1.1, 0.1), 2), fontsize=25)
ax.set(xlim=(-3.5*sigma,3.5*sigma),ylim=(0,1))
```

##### Properties of the Standard Normal Distribution:

<ul>
    <li>Cumulative area is close to 0 for $z$-scores close to $z = -3.49$.</li>
    <li>Cumulative area increases as the $z$-scores increase.</li>
    <li>Cumulative area for $z=0$ is $0.5000$.</li>
    <li>Cumulative area is close to 1 for $z$-scores close to $z=3.49$.</li>
</ul>

<b>How to use the Standard Normal Table (Table 4 from Appendix B):</b>

<ul>
    <li>The Standard Normal Table provides a list of $z$-scores rounded to <em>two</em> decimal places with its corresponding area (area to the <em>left</em> of the $z$-score).</li>
    <li>To find the area that corresponds to a particular $z$-score, locate the two left-most digits at the far left of the table.</li>
    <li>Once these digits have been located, then locate the last digit at the top of the table.</li>
    <li>The area (rounded to 4 decimal places) is the cross-section of these two numbers.</li>
    <li>If a $z$-score is not in the table, use the entry closest to that $z$-score.</li>
    <li>If a $z$-score is not in the table and midway between two $z$-scores, use the area midway between the corresponding areas.</li>
</ul>

##### Finding area to the left, right, and between $z$-scores

* Area to the left: just look up the area corresponding to the $z$-score
* Area to the right: $1 -($area corresponding to the $z$-score$)$
* Area between: find the area that corresponds to both $z$-socres and then subtract the smaller from the larger

##### Interpretation of Standard Normal Area:

* The Standard Normal Distribution is a <em>continuous probability distrubtion</em>.
* This means the area to the left of a given $z$-score represents the <em>probability</em> that a value is less than that $z$-score.
* Likewise, the area to the right of a given $z$-score represents the <em>probability</em> that a value is more than that $z$-score.
* The area between two given $z$-scores represents the <em>probability</em> that a value is between those $z$-scores.
* Values that satisfy $z < -2$ or $z > 2$ are considered unusual.
* Values that satisfy $z < -3$ or $z > 3$ are considered very unusual

##### Example 6

**(a)** Find the area to the right of $z = -0.65$.

**(b)** Find the area to the left of $z=-2.98$.

**(c)** Find the area to the left of $z = 1.365$.

**(d)** Find the area between $z = -1.96$ and $z = 1.96$.

**(e)** Find the total area: area to the left of $z=-1.28$ and to the right of $z = 1.28$.

**(f)** Find the total area: area to the left of $z = -1.53$ and to the right of $z = 0$.

```python
sigma = 1
mu = 0

x = np.arange(mu-4*sigma,mu+4*sigma+.1,.1)
y = P(x,sigma,mu)

z1 = -1.28
z2 = 1.28

x_fill_left = np.linspace(-10,z1,1000)
x_fill_right = np.linspace(z2,10,1000)
y_fill_left = P(x_fill_left,sigma,mu)
y_fill_right = P(x_fill_right,sigma,mu)

x_fill_between = np.linspace(z1,z2,1000)
y_fill_between = P(x_fill_between,sigma,mu)

fig, ax = plt.subplots(figsize=(15,7.5))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='blue',alpha=.05)

# #### area between
# ax.fill_between(x_fill_between,0,y_fill_between,color='red')

##### area to the left
ax.fill_between(x_fill_left,0,y_fill_left,color='red')

##### area to the right
ax.fill_between(x_fill_right,0,y_fill_right,color='red')

##### annotate left
ax.axvline(x=z1,ymin=0,ymax=2*P(z1,sigma,mu),color='k')
ax.annotate('$z={}$'.format(z1),xy=(z1+.05,0.025),xytext=(z1+0.05,0.025),color='k',fontsize=15)

##### annotate right
ax.axvline(x=z2,ymin=0,ymax=2*P(z2,sigma,mu),color='k')
ax.annotate('$z={}$'.format(z2),xy=(z2+.05,0.025),xytext=(z2+0.05,0.025),color='k',fontsize=15)

##### plot attributes
ax.set_title('Standard Normal Curve',fontsize=25)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_xticklabels(np.arange(mu-3*sigma,mu+3*sigma+1,sigma), fontsize=15)
ax.set_yticks(np.arange(0,1.1,.1))
ax.set_yticklabels(np.round(np.arange(0, 1.1, 0.1), 2), fontsize=15)
ax.set(xlim=(-4,4),ylim=(0,.5))
```
