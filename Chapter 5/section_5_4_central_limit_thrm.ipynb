##### Objective 1: Finding Sampling Distributions and Verify Their Properties

##### Sampling Distribution of Sample Means

* A <em><b>sampling distribution</b></em> is the probability distrubtion of a sample statistic that is formed when samples of size $n$ are repeatedly taken from a population

* If the sample statistic is the sample mean, then the distribution is the <em><b>sampling distribution of sample means</b></em>

* Every sample statistic has a sampling distribution.

##### Properties of Sampling Distributions of Sample Means:

* The mean of the sample means $\mu_{\bar{x}}$ is equal to the population mean $\mu$. In symbols, this means $\mu_{\bar{x}} = \mu$.

* The standard deviation of the sample means $\sigma_{\bar{x}}$ is equal to the population standard deviation $\sigma$ divided by the square root of the sample size $n$. In symbols, this means $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$.

##### Objective 2: Interpreting the Central Limit Theorem

##### Central Limit Theorem

* If samples of size $n$, where $n \ge 30$, are drawn from *any* population with a mean $\mu$ and a standard deviation $\sigma$, then the sampling distribution of sample means approximates a normal distribution. The greater the sample size, the better the approximation.

* If the population itself is normally distributed, then the sampling distribution of sample means is normally distributed for <em>any</em> sample size $n$.

* In either case, the following statements hold: $\mu_{\bar{x}} = \mu$ and $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

%matplotlib inline

import warnings

warnings.filterwarnings('ignore')

def f(x):
    return  np.sin(x) + np.random.normal(scale=0.1, size=len(x))

x1 = np.linspace(-2*np.pi,2*np.pi,70)
y1 = f(x1)

fig, (ax1, ax2) = plt.subplots(2,figsize=(15,10))
ax1.axhline(y=-1, color='k')
ax1.fill_between(x1,-1,y1,color='green')
ax1.plot(x1,y1,color='blue')
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set(xlim=(min(x1),max(x1)), ylim=(-1,max(y1)+.1))
ax1.set_title('Population Distribution (Not Normally Distributed)',fontsize=25)
ax1.set_xlabel('$x$',fontsize=25)

x2 = np.linspace(-5,5,100)
y2 = stats.norm.pdf(x2,0,1)

ax2.plot(x2,y2,color='blue')
ax2.axhline(y=0, color='k')
ax2.fill_between(x2,0,y2,color='green')
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set(xlim=(min(x2),max(x2)), ylim=(min(y2),max(y2)+.1))
ax2.set_title(r'Sampling Distribution of Sample Means for Population Distribution with $n \geq 30$',fontsize=25)
ax2.set_xlabel(r'$\bar{x}$',color='red',fontsize=25)
ax2.annotate(r'$\mu_{\bar{x}}$',(0.1,.25),color='blue',fontsize=25)
ax2.axvline(x=0,ymin=0,ymax=2*max(y2),color='k')
ax2.annotate(r'$\mu_{\bar{x}} + \sigma_{\bar{x}}$',(1.1,.125),color='blue',fontsize=25)
ax2.axvline(x=1,ymin=0,ymax=2*stats.norm.pdf(1,0,1),color='k')
ax2.annotate(r'$\mu_{\bar{x}} - \sigma_{\bar{x}}$',(-1.1,.125),color='blue',fontsize=25)
ax2.axvline(x=-1,ymin=0,ymax=2*stats.norm.pdf(-1,0,1),color='k')

fig.tight_layout()
```

```python
x1 = np.linspace(-5,5,100)
y1 = stats.norm.pdf(x1,0,.65)

fig, (ax1, ax2) = plt.subplots(2,figsize=(15,10))
ax1.axhline(y=-1, color='k')
ax1.fill_between(x1,0,y1,color='green')
ax1.plot(x1,y1,color='blue')
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set(xlim=(min(x1),max(x1)), ylim=(0,max(y1)+.1))
ax1.set_title('Population Distribution (Normally Distributed)',fontsize=25)
ax1.set_xlabel('$x$',fontsize=25)

x2 = np.linspace(-5,5,100)
y2 = stats.norm.pdf(x2,0,1)

ax2.plot(x2,y2,color='blue')
ax2.axhline(y=0, color='k')
ax2.fill_between(x2,0,y2,color='green')
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set(xlim=(min(x2),max(x2)), ylim=(min(y2),max(y2)+.1))
ax2.set_title(r'Sampling Distribution of Sample Means for Normally Distributed Population (Any $n$)',fontsize=25)
ax2.set_xlabel(r'$\bar{x}$',color='red',fontsize=25)
ax2.annotate(r'$\mu_{\bar{x}}$',(0.1,.25),color='blue',fontsize=25)
ax2.axvline(x=0,ymin=0,ymax=2*max(y2),color='k')
ax2.annotate(r'$\mu_{\bar{x}} + \sigma_{\bar{x}}$',(1.1,.125),color='blue',fontsize=25)
ax2.axvline(x=1,ymin=0,ymax=2*stats.norm.pdf(1,0,1),color='k')
ax2.annotate(r'$\mu_{\bar{x}} - \sigma_{\bar{x}}$',(-1.1,.125),color='blue',fontsize=25)
ax2.axvline(x=-1,ymin=0,ymax=2*stats.norm.pdf(-1,0,1),color='k')

fig.tight_layout()
```

Random samples of 100 cell phone bills are drawn from this population, and the mean of each sample is determined. Find the mean and standard deviation of the sampling distribution of sample means.

##### Solution:

* $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} = \frac{9}{\sqrt{100}} = 0.9$

##### Objective 3: Apply the Central Limit Theorem to Find the Probability of a Sample Mean

##### Finding the $z$-score for a Sampling Distribution of Sample Means

* We can find the probability that the sample mean $\bar{x}$ of a random sample will lie in a given interval of the $\bar{x}$ sampling distribution.

* We do this by transforming the sample mean $\bar{x}$ to a $z$-score, but now we must adjust our variables to agree with the sample's mean and standard deviation:

\begin{eqnarray}
z & = & \frac{\bar{x} - \mu_{\bar{x}}}{\sigma_{\bar{x}}} \\
z & = & \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
\end{eqnarray}

##### Example 4

The table below shows the lengths of time people spend driving each day. You randomly select 50 drivers ages 15 to 19. What is the probability that the mean time they spend driving each day is between 24.7 and 25.5 minutes? Assume that $\sigma = 1.5$ minutes.

<table style="width:50%">
    <tr> <th>Age Group</th> <th>Average Driving Time</th> </tr>
    <tr> <td>15-19</td> <td>25</td> </tr>
    <tr> <td>20-24</td> <td>52</td> </tr>
    <tr> <td>25-54</td> <td>64</td> </tr>
    <tr> <td>55-64</td> <td>58</td> </tr>
    <tr> <td>65+</td> <td>39</td> </tr>
</table>

##### Solution:

* In previous examples for finding probabilities that involve normal distributions, a key phrase such as, "assume that such and such is normally distributed" has been thrown in somewhere in the problem

* We don't have anything like that here... is this an issue?

* No! This is the beauty of the Central Limit Theorem!

* One of the conditions for the Central Limit Theorem is randomly selected samples of size $n \ge 30$

* We are randomly selecting a group of size $n=50$, which is certainly larger than 30

* Therefore, we can carry on using the modified formula for finding the $z$-score as determined above:

$$
    z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$

* Identifying all the necessary quantities in our formula:

    * We already determined that we are randomly selecting a group of size $n=50$

    * We also are given the assumption that $\sigma = 1.5$

    * From the table provided, we are interested in the 15-19 age group, of which the average driving time, or <em>mean</em> driving time, is $\mu = 25$

    * So the last thing we need to identify is $\bar{x}$

    * For this particular problem, since we are looking for the probability that the <em>mean of our sample of 50 drivers</em> in this age group is <em>between</em> two values, we will need <em>two</em> $x$-values: $\bar{x}_1 = 24.7$ and $\bar{x}_2 = 25.5$

* Using the modified formula from above we obtain two $z$-scores corresponding to each $x$-value

* For $\bar{x_1} = 24.7$:

$$
z_1 = \frac{\bar{x}_1 - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{24.7 - 25}{\frac{1.5}{\sqrt{50}}} \approx -1.41
$$

* For $\bar{x_2} = 25.5$:

$$
z_2 = \frac{\bar{x}_2 - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{25.5 - 25}{\frac{1.5}{\sqrt{50}}} \approx 2.36
$$

* Therefore,

\begin{eqnarray}
P(24.7 < \bar{x} < 25.5) & = & P(-1.41 < z < 2.36) \\
& = & 0.9909 - 0.0793 \\
& = & \boxed{0.9116}
\end{eqnarray}

* Interpreting this result:

    * If we randomly select 50 individuals aged 15-19 and find the average driving time of that sample, $\bar{x}$, there is a 91.16\% chance that this average driving time is between 24.7 minutes and 25.5 minutes.

##### Example 5

The mean room and board expense per year at four-year colleges is $\$9126$. You randomly select 9 four-year colleges. What is the probability that the mean room and board is less than $\$9400$? Assume that the room and board expenses are normally distributed with a standard deviation of $\$1500$.

##### Solution:

* What are we looking for?

* We are looking for $P(\bar{x} < 9400)$ for random samples of size $n = 9$

* Since we are looking for a probability that the <em>sample mean</em> is less than $\$9400$ and we are given that the distribution is normally distributed, we need to use the formula $z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}$

    * Note that the Central Limit Theorem applies regardless of the sampling size, $n$

* Noting that $\mu = 9126$, $\sigma = 1500$, and $n = 9$ we end up with

$$
    z = \frac{9400 - 9126}{\frac{1500}{\sqrt{9}}} \approx 0.55
$$

* Therefore, $P(\bar{x} < 9400) = P(z < 0.55) = \boxed{0.7088}$

* Interpreting this result:

    * If we randomly select 9 four-year colleges and find the average room and board expenses per year of that sample, $\bar{x}$, there is a 70.88\% chance that this average expense is less than \$9400.

##### Example 6

The average credit card debt carried by undergraduates is normally distributed, with a mean of $\$3173$ and a standard deviation of $\$1120$. **(a)** What is the probability that a randomly selected undergraduate, who is a credit card holder, has a credit card balance less than $\$2700$? **(b)** You randomly select 25 undergraduates who are credit card holders. What is the probability that their mean credit card balance is less than $\$2700$? **(c)** Compare the probabilities found in part **(a)** and **(b)**.

##### Solution:

**(a)** What are we looking for?

* We are looking for $P(x < 2700)$

    * But notice that we aren't looking for a <em>sample mean</em> here

    * We are looking for a <em>specific, individual</em> $x$-value

    * So we need to use the formula $z = \frac{x - \mu}{\sigma}$

* Using this formula with $\mu = 3173$ and $\sigma = 1120$ gives us

$$
    z = \frac{2700 - 3173}{1120} \approx -0.42
$$

* Therefore, we have $P(x < 2700) = P(z < -0.42) = \boxed{0.3372}$ or about a $33.72\%$ chance that $x < \$2700$

**(b)** What are we looking for?

* We're looking for $P(\bar{x} < 2700)$ for random samples of size $n = 25$.

    * Since we are dealing with random samples and we are given that the distribution is normally distributed, we need to use the formula $z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}$
    
        * Central Limit Theorem applies regardless of the size of the sampling size, $n$

* Therefore,

$$
    z = \frac{2700 - 3173}{\frac{1120}{\sqrt{25}}}  \approx -2.11
$$

* So we have $P(\bar{x} < 2700) = P(z < -2.11) = \boxed{0.0174}$ or about a $1.74\%$ chance that $\bar{x} < \$2700$

**(c)**

* Comparing the results from each previous part, we can clearly see that the probability in scenario **(a)** is much higher than that of scenario **(b)**

* In other words, it is much more likely that we find a <em>single</em> person that has a balance below $\$2700$ than finding a random sample of size 25 that has a sample mean, $\bar{x}$, that is below $\$2700$
