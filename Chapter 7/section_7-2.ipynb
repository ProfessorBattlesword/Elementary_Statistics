##### Objective 1: Finding and Interpreting $P$-values

##### Using a $P$-value to Make a Decision

To use a $P$-value to make a decision in a hypothesis test, compare the $P$-value with $\alpha$.

If $P \leq \alpha$, then reject $H_0$.

If $P > \alpha$, then do not reject $H_0$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

%matplotlib inline

sigma = 1
mu = 0

x = np.arange(mu-4*sigma,mu+4*sigma+.01,.01)
y = stats.norm.pdf(x,mu,sigma)

z1 = -1.28
z2 = 1.28

x_fill_left = np.linspace(-10,z1,1000)
x_fill_right = np.linspace(z2,10,1000)
y_fill_left = stats.norm.pdf(x_fill_left,mu,sigma)
y_fill_right = stats.norm.pdf(x_fill_right,mu,sigma)

x_fill_between = np.linspace(z1,z2,1000)
y_fill_between = stats.norm.pdf(x_fill_between,mu,sigma)

fig, ax = plt.subplots(figsize=(15,7.5))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='blue',alpha=.05)
# ax.fill_between(x_fill_between,0,y_fill_between,color='red') ##### AREA BETWEEN
ax.fill_between(x_fill_left,0,y_fill_left,color='red') ##### AREA TO LEFT
ax.fill_between(x_fill_right,0,y_fill_right,color='red') ##### AREA TO THE RIGHT
ax.set_title('Standard Normal Curve',fontsize=15)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_yticks(np.arange(0,stats.norm.pdf(mu,mu,sigma)+.1,.1))
ax.set(xlim=(min(x),max(x)),ylim=(0,max(y)+.1))
ax.annotate(r'$A_{left}=\frac{1}{2}P$',xy=(z1+.05,0.025),xytext=(z1-0.75,0.025),color='k',fontsize=15)
ax.annotate(r'$A_{right}=\frac{1}{2}P$',xy=(z2+.05,0.025),xytext=(z2+0.05,0.025),color='k',fontsize=15)
```

##### Example:

The $P$-value for a hypothesis test is $P = 0.0237$. What is your decision when the level of significance is (a) $\alpha = 0.05$ and (b) $\alpha = 0.01$.

##### Solution:

(a) Since $P = 0.0237 < 0.05 = \alpha$, we would reject whatever the null hypothesis happens to be. So our decision in this instance would be to reject $H_0$.

(b) Since now we have $P = 0.0237 > 0.01 = \alpha$, we would fail to reject the null hypothesis. So in this case we would fail to reject $H_0$.

##### Finding $P$-values

After determining the hypothesis test's standardized test statistic and the standardized test statistic's corresponding area, we do one of the following to find the $P$-value:

* For a left tailed test ($H_a$ contains $<$), $P =$ Area in left tail.
* For a right tailed test ($H_a$ contains $>$), $P =$ Area in the right tail.
* For a two-tailed test ($H_a$ contains $\neq$), then $P =$ Twice the area in either tail of the standardized test statistic.

##### Example:

Find the $P$-value for a left-tailed hypothesis test with a standardized test statistic of $z = -2.23$. Decide whether to reject $H_0$ when the level of significance is $\alpha = 0.01$.

##### Solution:

We need to find the area to the left of $z = -2.23$. The standard normal table tells us that $P = 0.0129$. Since $P = 0.0129 > 0.01 = \alpha \Rightarrow$ fail to reject $H_0$.

##### Example:

Find the $P$-value for a right-tailed hypothesis test with a standardized test statistic of $z=1.71$. Decide whether to reject $H_0$ when the level of significance is $\alpha = 0.05$.

##### Solution:

The area to the right of $z = 1.71$ is $P = 1 - 0.9564 = 0.0436 < 0.05 \Rightarrow$ reject $H_0$

##### Example:

Find the $P$-value for a two-tailed hypothesis test with a standardized test statistic of $z=2.14$. Decide whether to reject $H_0$ when the level of significance is $\alpha = 0.05$.

##### Solution:

For a two-tailed test we are looking at the area to the left of $-z$ and to the right of $z$. Since the standard normal curve is symmetric these area are the same.

The area to the left of $-z = -2.14$ is $\frac{1}{2}P = 0.0162$. Notice that this is <em>half</em> of $P$! We need the <em>combined area in both tails</em>. So we multiply by two and get $P = 2(0.0162) = 0.0328 < 0.05 = \alpha \Rightarrow$ reject $H_0$.

##### Objective 2: Using $P$-values for a $z$-test for a Mean ($\sigma$ Known)

##### $z$-Test for a Mean $\mu$

Similar to confidence intervals, if the population standard deviation ($\sigma$) is known, then we can use the following formula to transform a sample mean into a $z$-score: $$z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}$$

This, of course, only true if we know the sample is random and the Central Limit Thoerem applies (either $n \geq 30$ or the population is normally distributed).

##### Steps for Performing a Hypothesis Test when the Population Standard Deviation ($\sigma$) is Known

1. Verify that $\sigma$ is known, the sample is random, and either $n \geq 30$ or the population is normally distributed.
2. State the claim mathematically and verbally and identify the null and alternative hypotheses.
3. Specify the level of significance (identify $\alpha$).
4. Find the standardized test statistic $\left( z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} \right)$.
5. Find the area that corresponds to $z$ (use standard normal table/calculator; also depends on what tail test you have).
6. Find the $P$-value (depends on what tail test you have).
7. Make a decision to reject or fail to reject the null hypothesis $\left( P \leq \alpha \Rightarrow \text{ reject } H_0; P > \alpha \Rightarrow \text{ fail to reject } H_0 \right)$.
8. Interpret the decision in the context of the original claim.

##### Example:

In auto racing, a pit stop is where a racing vehicle stops for new tires, fuel, repairs, and other mechanical adjustments. The efficiency of a pit crew that makes these adjustments can affect the outcome of a race. A pit crew claims that its mean pit stop time (for 4 new tires and fuel) is less than 13 seconds. A random sample of 32 pit stop times has a sample mean of 12.9 seconds. Assume the population standard deviation is 0.19 seconds. Is there enough evidence to support the claim at $\alpha = 0.01$? Use a $P$-value to make the decision.

##### Solution:

1. We sample standard deviation is given so $\sigma = 0.19$. We have a random sample and the sample size is $n = 32 \geq 30$ so the Central Limit Theorem applies.

2. The null and alternative hypotheses with the claim expressed mathematically is

\begin{eqnarray}
\nonumber H_0: \mu & \geq & 13 \\ \nonumber H_a: \mu & < & 13 \; \text{(claim)}
\end{eqnarray}

Since $H_a$ contains $<$, we are dealing with a left-tailed test.

3. The level of significance is given by $\alpha = 0.01$.

4. Noting that $\bar{x} = 12.9$ sec, the standardized test statistic is calculated as

$$z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{12.9 - 13}{\frac{0.19}{\sqrt{32}}} \approx -2.98$$

5. Since we are dealing with a left-tailed test, we find the area to the left of this $z$ value:

$$A_{left} = 0.0014$$

6. Again, noting that we have a left-tailed test, we have $P = 0.0014$.

7. To make a decision we compare the $P$-value with $\alpha$: $P = 0.0014 < 0.01 = \alpha \Rightarrow$ reject $H_0$.

8. Interpretation: There is <u>enough</u> evidence at the $1\%$ level of significance to <u>support</u> the claim that the mean pit stop time is less than 13 seconds (it <em>appears</em> the pit crew is correct in their claim).

##### Example:

According to a study the mean cost of bariatric (wheight loss) surgery is $\$21500$. You think this information is incorrect. You randomly select 25 bariatric surgery patients and find that the mean cost for their surgeries is $\$20695$. From past studies, the population standard deviation is known to be $\$2250$ and the population is normally distributed. Is there enough evidence to support <em>your claim</em> at $\alpha = 0.05$? Use a $P$-value test.

##### Solution:

1. We are given the population standard deviation from previous studies as $\sigma = 2250$. We have $n=25 < 30$, however, we are also given that the population is known to be normally distributed, so the Central Limit Theorem applies.

2. The null and alternative hypotheses with the claim expressed mathematically is

\begin{eqnarray}
\nonumber H_0: \mu & = & 21500 \\ \nonumber H_a: \mu & \neq & 21500 \; \text{(claim)}
\end{eqnarray}

Since $H_a$ contains $\neq$, we are dealing with a two-tailed test.

3. The level of significance is given by $\alpha = 0.05$.

4. Noting that $\bar{x} = 20695$, the standardized test statistic is calculated as

$$z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{20695 - 21500}{\frac{2250}{\sqrt{25}}} \approx -1.79$$

5. Since we are dealing with a two-tailed test, we find the area to the left of this $z$ value (since it's negative; if z is positive in two-tailed test, then find the area to the right):

$$A_{left} = 0.0367$$

6. Again, noting that we have a two-tailed test, we have $P = 2(0.0367) = 0.0734$ (multiplying by 2 because we have two-tailed test).

7. To make a decision we compare the $P$-value with $\alpha$: $P = 0.0734 > 0.05 = \alpha \Rightarrow$ do not reject $H_0$.

8. Interpretation: There is <u>not enough</u> evidence at the $5\%$ level of significance to <u>support</u> the claim that the mean cost of bariatric surgery is different than $\$21500$ (it <em>appears</em> that <em>your</em> claim is incorrect).

##### Objective 3: Finding Critical Values and Rejection Regions in the Standard Normal Distribution

##### Definitions:

A <em><b>rejection region (critical region)</b></em> of a sampling distribution is the range of values for which the null hypothesis is not probable.

A <em><b>critical value</b></em>, $z_0$, separates the rejection region from the nonrejection region.

If a standardized test statistic is within the rejection region, then the null hypothesis is rejected.

##### Finding Critical values in the Standard Normal Distribution

* Specify the level of significance, $\alpha$.
* Determine whether the test is left-tailed, right-tailed, or two-tailed.
* Find the critical value(s), $z_0$.
  * Left-tailed test $\Rightarrow$ find the $z$-score that corresponds to an area equal to $\alpha$
  * Right-tailed test $\Rightarrow$ find the $z$-score that corresponds to an area of $1-\alpha$
  * Two-tailed test $\Rightarrow$ Find the $z$-score that corresponds to $\frac{1}{2}\alpha$ and $1-\frac{1}{2}\alpha$

##### Example:

Find the critical value and rejection region for a right-tailed test with $\alpha = 0.01$.

##### Solution:

We have a right-tailed test and $\alpha = 0.01 \Rightarrow A_{right} = 1 - \alpha = 0.99$.

This right-tailed area corresponds to $\boxed{z_0 = 2.33}$. The rejection region is any $z$-value <em>larger</em> or <em>to the right</em> of this critical value.

##### Example:

Find the critical values and rejection regions for a two-tailed test with $\alpha = 0.05$.

##### Solution:

Since we have a two-tailed test and $\alpha = 0.05$, we have a left and right tail to consider:

$A_{left} = \alpha / 2 = 0.025$ and $A_{right} = 1 - \alpha / 2 = 0.025$.

Really, we need only find one of these $z$-values (since the standard normal distribution is symmetric) and the other value will be the opposite sign. So we'll just look for the corresponding $z$-value for the area in the left tail:

$$-z_0 = -1.96$$

Since the value is negative (left critical value), we know that this $z$-value is $-z_0$. This means that our positive critical value (right critical value) is $z_0 = 1.96$.

So our two critical values are $\boxed{-z_0 = -1.96 \text{ and } z_0 = 1.96}$. And the rejection regions are any $z$-values that lie to the <em>left</em> of $-z_0 = -1.96$ and any $z$-values that lie to the <em>right</em> of $z_0 = 1.96$.

##### Objective 4: Using Rejection Regions for a $z$-test for a Mean ($\sigma$ Known)

##### Using Rejection Regions for a $z$-Test for a Mean ($\sigma$ Known)

1. verify that $\sigma$ is known, the sample is random, and either $n \geq 30$ or the population is normally distributed.
2. State the claim mathematically and verbally and identify the null and alternative hypothees.
3. Specify the level of significance (identify $\alpha$).
4. Determine the critical value(s) (use standard normal table to find $z$-scores corresponding to appropriate area(s)).
5. Determine the rejection region(s).
6. Find the standardized test statistic $\left( z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} \right)$.
7. Make a decision to reject or fail to reject the null hypotheis (if $z$ is in the rejection region, then reject $H_0$; otherwise fail to reject $H_0$).
8. Interpret the decision in the context of the original claim.

##### Example:

Employees at a construction and mining company claim that the mean salary of the company's mechanical engineers is less than that of one of its competitors, which is $\$68000$. A random sample of 20 of the company's mechanical engineers has a mean salaray of $\$66900$. Assume the population standard deviation is $\$5500$ and the population is normally distributed. At $\alpha = 0.05$, test the employees' claim.

##### Solution:

1. We are to assume the population standard deviation is $\$5500$ ($\sigma = 5500$). We have a <em>random</em> sample of only $n = 20$ mechanical engineers, but we are also to assume the population is normally distributed so the Central Limit Theorem applies.

2. The hypotheses are

\begin{eqnarray}
\nonumber H_0: \mu & \geq & 68000 \\
\nonumber H_a: \mu & < & 68000 \; \text{(claim)}
\end{eqnarray}

So this is a left-tailed test.

3. The level of significance is given: $\alpha = 0.05$.

4. If we use the standard normal table, we notice that $\alpha = 0.05$ is not listed as an area. Rather, this value lies precisely in the middle between two values that are listed ($0.0495$ and $0.0505$), so we must average the corresponding $z$-values for that which is listed.

The critical value for a left-tailed test with $\alpha = 0.05$ is:

$$z_0 = -1.645$$



5. Since we are dealing with a left-tailed test the rejection region is all $z$ values to the left (less than) $z_0 = -1.645$.

Summarizing:

$$\text{Rejection region is: } z < -1.645$$

6. Now we calculate the standardized test statistic noting that our <em>sample</em> has a mean of $\bar{x} = 66900$:

$$z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{66900-68000}{\frac{5500}{\sqrt{20}}} \approx -0.89$$

7. Since $z = -0.89 > -1.645 = z_0$, we <em>do not</em> reject $H_0$.

8. Interpretation: There is <u>not enough</u> evidence at the $5\%$ level of significance to <u>support</u> the employer's claim that the mean salary of mechanical engineers is less than $\$68000$.

##### Example:

A researcher claims that the mean annual cost of rasing a child (age 2 and under) by husband-wife families in the U.S. is $\$13960$. In a random sample of husband-wife families in the U.S., the mean annual cost of rasing a child (age 2 and under) is $\$13725$. The sample consists of $500$ children. Assume the population standard deviation is $\$2345$. At $\alpha = 0.10$, is there enough evidence to reject the claim?

##### Solution:

1. We are told that the population standard deviation is $\sigma = 2345$. We have a sample size of $n = 500 \geq 30$ and the sample is <em>random</em> so the Central Limit Theorem applies.

2. The hypotheses are

\begin{eqnarray}
\nonumber H_0: \mu & = & 13960 \; \text{(claim)}\\
\nonumber H_a: \mu & \neq & 13960
\end{eqnarray}

So this is a two-tailed test.

3. The level of significance is given: $\alpha = 0.10$.

4. Note that since this is a <em>two-tailed test</em> we will have <em>two critical values</em> ($-z_0$ and $z_0$). We find these critical values by first finding $-z_0$ and then, using symmetry, noting that $z_0$ is just the opposite sign.

To find $-z_0$ we look for $A_{left} = \alpha / 2 = 0.10 / 2 = 0.05 \Rightarrow -z_0 = -1.645$.

The $\alpha /2$ in the left tail is again because we are dealing with a two-tailed test.

5. Since we are dealing with a two-tailed test the rejection regions are all $z$ values to the left (less than) $-z_0 = -1.645$ and all $z$ values to the right (more than) $z_0 = 1.645$. So if our standardized test statistic is in <em>either</em> region we reject $H_0$. If our standardized test statistic is <em>between</em> these critical numbers, we <em>do not</em> reject $H_0$.

Summarizing:

$$\text{Rejection regions are: } z < -1.645 \text{ and } z > 1.645$$

6. Now we calculate the standardized test statistic noting that our <em>sample</em> has a mean of $\bar{x} = 13725$:

$$z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{13725-13960}{\frac{2345}{\sqrt{500}}} \approx -2.24$$

7. Since $z = -2.24 < -1.645 = z_0$, we <em>reject</em> $H_0$.

8. Interpretation: There is <u>enough</u> evidence at the $10\%$ level of significance to <u>reject</u> the researcher's claim that the mean cost of raising a child (age 2 and under) is $\$13960$.

In this case, based on our statistical analysis, it appears that the cost is <em>lower</em> than the claim asserts.

But what about other values for our standardized test statistc, $z$?

If we had found a <em>positive</em> $z$-value that was in the other rejection region, then we would still reject the claim, but this would be because it would appear that the cost is <em>more</em> than the claim asserts.

Yet again, if we had found a $z$-value that was <em>not</em> in either rejection region, then we would have sided with the researcher.

In <em>each</em> of these secenarios, we could still be wrong. We could make a type I or type II error. Just keep this in mind whenever you're doing hypothesis testing. Just because we <em>have reason to believe</em> a claim is correct or incorrect based on our statistical analysis, does not guarentee that we are correct in our decision. We are, however, <em>likely</em> to be correct.
