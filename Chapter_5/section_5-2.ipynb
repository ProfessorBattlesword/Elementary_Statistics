# Chapter 5: Normal Probability Distributions

### Sections 5.2: Finding Probabilities Using Standard Normal Distribution

##### Objective 1: Finding Probabilities for Normally Distributed Variables

##### Recall the Standard Score: $z = \frac{x-\mu}{\sigma}$

Interpretation: given a specific value $x$ from a population data set with population mean $\mu$ and population standard deviation $\sigma$, we can transform $x$ into a $z$-score (standard score) and use the Standard Normal Distribution to find probabilities involving $x$.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

%matplotlib inline

sigma1 = 50
mu1 = 350
sigma2 = 1
mu2 = 0

x1 = np.arange(mu1-4*sigma1,mu1+4*sigma1+.1,.1)
y1 = stats.norm.pdf(x1,mu1,sigma1)

x2 = np.arange(mu2-4*sigma2,mu2+4*sigma2+.1,.1)
y2 = stats.norm.pdf(x2,mu2,sigma2)

fig, (ax1,ax2) = plt.subplots(2,figsize=(15,10))
ax1.plot(x1, y1,color='blue')
ax1.grid(True, which='both')
ax1.axhline(y=0, color='k')
ax1.axvline(x=0, color='k')
ax1.fill_between(x1,0,y1,color='green')
ax1.set_title('Normal Curve')
ax1.set_xticks(np.arange(mu1-3*sigma1,mu1+3*sigma1+1,sigma1))
ax1.set_xticklabels(np.arange(mu1-3*sigma1,mu1+3*sigma1+1,sigma1),fontsize=15)
ax1.set(xlim=(mu1-3*sigma1,mu1+3*sigma1),ylim=(0,.01))
ax1.set_xlabel('$x$',fontsize=25)

ax2.plot(x2, y2,color='blue')
ax2.grid(True, which='both')
ax2.axhline(y=0, color='k')
ax2.axvline(x=0, color='k')
ax2.fill_between(x2,0,y2,color='green')
ax2.set_title('Normal Curve')
ax2.set_xticks(np.arange(mu2-3*sigma2,mu2+3*sigma2+1,sigma2))
ax2.set_xticklabels(np.arange(mu2-3*sigma2,mu2+3*sigma2+1,sigma2),fontsize=15)
ax2.set(xlim=(mu2-3*sigma2,mu2+3*sigma2),ylim=(0,.5))
ax2.set_xlabel('$z$',fontsize=25)
ax2.set_ylabel('Probability Density',fontsize=25)

fig.tight_layout()
```

##### Example:

A survey indicates that people keep their cell phone an average of 1.5 years before buying a new one. The standard deviation is 0.25 years. A cell phone user is selected at random. Find the probability that the user will keep his or her current phone for less than 1 year before buying a new one. Assume that the lengths of time people keep their phone are normally distributed and are represented by the variable $x$.

##### Solution:

Keep in mind we need to convert from the given values into the Standard Normal distribution; in pictures:

```python
sigma1 = .25
mu1 = 1.5
sigma2 = 1
mu2 = 0

x1 = np.arange(mu1-4*sigma1,mu1+4*sigma1+.1,.01)
y1 = stats.norm.pdf(x1,mu1,sigma1)

x2 = np.arange(mu2-4*sigma2,mu2+4*sigma2+.1,.1)
y2 = stats.norm.pdf(x2,mu2,sigma2)

fig, (ax1,ax2) = plt.subplots(2,figsize=(15,10))
ax1.plot(x1, y1,color='blue')
ax1.grid(True, which='both')
ax1.axhline(y=0, color='k')
ax1.axvline(x=0, color='k')
ax1.fill_between(x1,0,y1,color='green')
ax1.set_title('Normal Curve')
ax1.set_xticks(np.arange(mu1-3*sigma1,mu1+3*sigma1+1,sigma1))
ax1.set_xticklabels(np.arange(mu1-3*sigma1,mu1+3*sigma1+1,sigma1),fontsize=15)
ax1.set(xlim=(mu1-3*sigma1,mu1+3*sigma1),ylim=(0,2))
ax1.set_xlabel('Number of years a person keeps their cell phone ($x$)',fontsize=25)

ax2.plot(x2, y2,color='blue')
ax2.grid(True, which='both')
ax2.axhline(y=0, color='k')
ax2.axvline(x=0, color='k')
ax2.fill_between(x2,0,y2,color='green')
ax2.set_title('Normal Curve')
ax2.set_xticks(np.arange(mu2-3*sigma2,mu2+3*sigma2+1,sigma2))
ax2.set_xticklabels(np.arange(mu2-3*sigma2,mu2+3*sigma2+1,sigma2),fontsize=15)
ax2.set(xlim=(mu2-3*sigma2,mu2+3*sigma2),ylim=(0,.5))
ax2.set_xlabel('$z$',fontsize=25)
ax2.set_ylabel('Probability Density',fontsize=25)

fig.tight_layout()
```

So this is effectively how we can think about it: we convert the given distribution into the Standard Normal distribution. However, we don't actually need to do this for all possible values. Really we're just after the <em>probability</em> that a random person keeps their phone for <em>less than a year</em>.

This means we are looking for $P(x < 1)$. And we can convert $x$-values using $z = \frac{x - \mu}{\sigma}$. So we need to convert $x = 1$ year into a value represented by the Standard Normal distribution. This gives us $z = \frac{1-1.5}{0.25} = -2$.

Therefore, $P(x < 1) = P(z < -2)$.

Using a table or a calculator we can find $P(z < -2)$ by noting that we want the area to the left:

```python
sigma = .25
mu = 1.5

x = np.arange(mu-4*sigma,mu+4*sigma+.01,.01)
y = stats.norm.pdf(x,mu,sigma)

z1 = 1
z2 = 2

x_fill_left = np.linspace(-10,z1,1000)
x_fill_right = np.linspace(z2,10,1000)
y_fill_left = stats.norm.pdf(x_fill_left,mu,sigma)
y_fill_right = stats.norm.pdf(x_fill_right,mu,sigma)

x_fill_between = np.linspace(z1,z2,1000)
y_fill_between = stats.norm.pdf(x_fill_between,mu,sigma)

fig, ax = plt.subplots(figsize=(15,7.5))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='blue',alpha=.05)
# ax.fill_between(x_fill_between,0,y_fill_between,color='red') ##### AREA BETWEEN
ax.fill_between(x_fill_left,0,y_fill_left,color='red') ##### AREA TO LEFT
# ax.fill_between(x_fill_right,0,y_fill_right,color='red') ##### AREA TO THE RIGHT
ax.set_title('Starting Normal Curve',fontsize=15)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_yticks(np.arange(0,stats.norm.pdf(mu,mu,sigma)+.1,.1))
ax.set(xlim=(min(x),max(x)),ylim=(0,max(y)+.1))
ax.axvline(x=z1,ymin=0,ymax=.5*stats.norm.pdf(z1,mu,sigma),color='k') ##### ANNOTATE LEFT
ax.annotate('$x={}$'.format(z1),xy=(z1+.05,0.025),xytext=(z1+0.05,0.025),color='k',fontsize=15)
# ax.axvline(x=z2,ymin=0,ymax=0.5*stats.norm.pdf(z2,mu,sigma),color='k') ##### ANNOTATE RIGHT
# ax.annotate('$z={}$'.format(z2),xy=(z2+.05,0.025),xytext=(z2+0.05,0.025),color='k',fontsize=15)
```

The Standard Normal equivalence:

```python
sigma = 1
mu = 0

x = np.arange(mu-4*sigma,mu+4*sigma+.01,.01)
y = stats.norm.pdf(x,mu,sigma)

z1 = -2
z2 = 2

x_fill_left = np.linspace(-10,z1,1000)
x_fill_right = np.linspace(z2,10,1000)
y_fill_left = stats.norm.pdf(x_fill_left,mu,sigma)
y_fill_right = stats.norm.pdf(x_fill_right,mu,sigma)

x_fill_between = np.linspace(z1,z2,1000)
y_fill_between = stats.norm.pdf(x_fill_between,mu,sigma)

fig, ax = plt.subplots(figsize=(15,7.5))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='blue',alpha=.05)
# ax.fill_between(x_fill_between,0,y_fill_between,color='red') ##### AREA BETWEEN
ax.fill_between(x_fill_left,0,y_fill_left,color='red') ##### AREA TO LEFT
# ax.fill_between(x_fill_right,0,y_fill_right,color='red') ##### AREA TO THE RIGHT
ax.set_title('Standard Normal Curve',fontsize=15)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_yticks(np.arange(0,stats.norm.pdf(mu,mu,sigma)+.1,.1))
ax.set(xlim=(min(x),max(x)),ylim=(0,max(y)+.1))
ax.axvline(x=z1,ymin=0,ymax=.5*stats.norm.pdf(z1,mu,sigma),color='k') ##### ANNOTATE LEFT
ax.annotate('$z={}$'.format(z1),xy=(z1+.05,0.025),xytext=(z1+0.05,0.025),color='k',fontsize=15)
# ax.axvline(x=z2,ymin=0,ymax=0.5*stats.norm.pdf(z2,mu,sigma),color='k') ##### ANNOTATE RIGHT
# ax.annotate('$z={}$'.format(z2),xy=(z2+.05,0.025),xytext=(z2+0.05,0.025),color='k',fontsize=15)
```

From the table: $P(x < 1) = P(z < -2) \approx 0.0228$.

So that means about 2.28\% of cell phone users will keep their phone for less than 1 year before buying a new one.

##### Example:

The average speed of vehicles traveling on a stretch of highway is 67 miles per hour with a standard deviation of 3.5 miles per hour. A vehicle is selected at random. What is the probability that it is violating the speed limit of 70 miles per hour? Assume the speeds are normally distributed and are represented by the variable $x$.

##### Solution:

What are we looking for?

We need to find the <em>probability</em> that a randomly selected vehicle is <em>violating</em> the speed limit. If a vehicle is violating the speed limit... then that vehicle must be going <em>faster</em> ($x$ is larger) than the posted limit.

Therefore, we are looking to find $P(x > 70)$. Using a similar approach to the example before we have $z = \frac{x-\mu}{\sigma} = \frac{70-67}{3.5} \approx 0.86$ ($z$-scores we generally round to <em>two</em> decimal places). $P(x > 70) = P(z > 0.86)$.

To find the area that corresponds to this probability, we need to find the area to the <em>right</em> of the $z$-score (because of the greater than).

In a picture:

```python
sigma = 1
mu = 0

x = np.arange(mu-4*sigma,mu+4*sigma+.01,.01)
y = stats.norm.pdf(x,mu,sigma)

z1 = 1
z2 = 0.86

x_fill_left = np.linspace(-10,z1,1000)
x_fill_right = np.linspace(z2,10,1000)
y_fill_left = stats.norm.pdf(x_fill_left,mu,sigma)
y_fill_right = stats.norm.pdf(x_fill_right,mu,sigma)

x_fill_between = np.linspace(z1,z2,1000)
y_fill_between = stats.norm.pdf(x_fill_between,mu,sigma)

fig, ax = plt.subplots(figsize=(15,7.5))
ax.plot(x, y,color='blue')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.fill_between(x,0,y,color='blue',alpha=.05)
# ax.fill_between(x_fill_between,0,y_fill_between,color='red') ##### AREA BETWEEN
# ax.fill_between(x_fill_left,0,y_fill_left,color='red') ##### AREA TO LEFT
ax.fill_between(x_fill_right,0,y_fill_right,color='red') ##### AREA TO THE RIGHT
ax.set_title('Standard Normal Curve',fontsize=15)
ax.set_xticks(np.arange(mu-3*sigma,mu+3*sigma+1,sigma))
ax.set_yticks(np.arange(0,stats.norm.pdf(mu,mu,sigma)+.1,.1))
ax.set(xlim=(min(x),max(x)),ylim=(0,max(y)+.1))
# ax.axvline(x=z1,ymin=0,ymax=.5*stats.norm.pdf(z1,mu,sigma),color='k') ##### ANNOTATE LEFT
# ax.annotate('$z={}$'.format(z1),xy=(z1+.05,0.025),xytext=(z1+0.05,0.025),color='k',fontsize=15)
ax.axvline(x=z2,ymin=0,ymax=0.5*stats.norm.pdf(z2,mu,sigma),color='k') ##### ANNOTATE RIGHT
ax.annotate('$z={}$'.format(z2),xy=(z2+.05,0.025),xytext=(z2+0.05,0.025),color='k',fontsize=15)
```

So we have $P(x>70) = P(z>0.86) \approx 0.1949$ (again, either using a table or calculator).

Interpreting this answer: there is about a 19.49\% (nearly 20\%) chance that a randomly selected car is speeding. That means we can expect about 1 in 5 cars to be speeding on this particular stretch of highway.

##### Example:

A survey indicates that for each trip to a supermarket, a shopper spends an average of 45 minutes with a standard deviation of 12 minutes in the store. The lengths of time spent in the store are normally distributed and are represented by the variable $x$. A shopper enters the store. (a) Find the probability that the shopper will be in the store for each interval of time listed below. (b) Interpret your answer when 200 shoppers enter the store. How many shoppers would you expect to be in the store for each interval of time listed below?

1. Between 24 and 54 minutes
2. More than 39 minutes

##### Solution:

For either case, to find part (a) we need to use $z = \frac{x-\mu}{\sigma}$ to convert our given $x$ values into $z$-scores. And then, according to each case, we find the appropriate area (probability). Note here that we are given an <em>average</em> (mean) of $\mu = 45$ minutes and a standard deviation of $\sigma = 12$ minutes spent in the store.

For part (b) we simply need to multiply the probability found in part (a) by 200. This tells us the <em>expected</em> number of customers that spend time in the store according to each case.

For case 1 part (a) we actually have two $x$-values: $x_1 = 24$ and $x_2 = 54$.

We convert these to $z$-scores: $z = \frac{x-\mu}{\sigma} \Rightarrow z_1 = \frac{24-45}{12} \approx -1.75$ and $z_2 = \frac{54-45}{12} \approx 0.75$.

And we wish for our random $x$ value to be <em>between</em> these values (this means we need to find the area between these values).

So we're looking for $P(24<x<54) = P(-1.75<z<0.75)$.

1. (a) $P(24<x<54) = (-1.75<z<0.75) \approx 0.7333 \Rightarrow$ for 200 shoppers (b) about $0.7333\cdot 200\approx 147$ would be expected to spend between 24 and 54 minutes in the store.

For case 2 part (a) we just need to convert one $x$-value: $x = 39$. This gives us a $z$-score of $z = \frac{x-\mu}{\sigma} = \frac{39-45}{12} \approx -0.50$. Here we note that we are looking for a random value that is <em>more than</em> 39 minutes (this indicates area to the right).

So we're looking for $P(x>39) = P(z>-0.50)$.

2. (a) $P(x>39) = P(z>-0.5) \approx 0.6915 \Rightarrow$ for 200 shoppers (b) about $0.6915 \cdot 200 \approx 138$ would be expected to spend more than 39 minutes in the store.
